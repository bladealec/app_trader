# App Picture Scraper
As of now, this script is only focused on scraping application names that are in enlgish and from the google store.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
```

#### retrieving .csv as a dataframe

```python
name_union = pd.read_csv('../app_trader_data/distinct_play_store.csv')
name_union
```

```python
name_list = name_union.values.tolist()
name_list
```

#### connecting to URL
- the URL template is https://play.google.com/store/search?q=**app_name**&c=apps
- spaces in the app name should be replaced with '+'

```python
URL = 'https://play.google.com/store/search?q=san+andreas+crime+stories+game&c=apps'
response = requests.get(URL)
```

```python
response.status_code
```

#### reading page elements

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

#### finding the link to the app's picture

```python
images = soup.findAll('img')
print(type(images))
images
```

```python
picture_link = soup.findAll('div', attrs={'class' : 'Shbxxd'})[0].find('img').get('src')
print(picture_link)
```
